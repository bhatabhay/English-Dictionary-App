```python
while True:
    print('MAIN MENU'.center(100, '*'))
    print('Add a new word (press 1)')
    print('Find the meaning (press 2)')
    print('Update a word (press 3)')
    print('Exit (press 0)')
    value = int(input('Enter the choice : '))
    if value == 0 :
        break
    elif value == 2:
        word = input('Enter the word : ').capitalize()
        with open('dictionary.txt', 'rb') as f:
            words = pickle.load(f)
            if word in words:
                print(words[word])
            else:
                print('Word not found!!')
    elif value == 1:
        word_meaning = input('Enter the word followed by its meaning separated by a - ').split('-')
        word = word_meaning[0].capitalize()
        meaning = word_meaning[1].capitalize()
        with open('dictionary.txt', 'r+b') as f:
            pickle.dump({word:meaning}, f)
            f.seek(0)
            words = pickle.load(f)
            print(words[word])
    elif value == 3:
        
        word_meaning = input('Enter the word followed by its meaning separated by a "-" ').split('-')
        word = word_meaning[0].capitalize()
        meaning = word_meaning[1].capitalize()
        with open('dictionary.txt', 'r+b') as f:
            pickle.dump({word:meaning}, f)
            f.seek(0)
            words = pickle.load(f)
            print(word,'-',words[word])
    else:
        print('Please enter right choice.....')
        
            
        
        
```
